# Preprocessing data
---



```python
#Melihat missing value pada data
df.isnull().sum()
```

Terdapat keanehan pada jumlah missing value yang sangat banyak bahkan hampir semuanya terdapat missing value, dicurigai terdapat baris kosong pada data

```python
#Menghilangkan baris data yang memiliki data kosong lebih dari 3 kolom
df.dropna(thresh = 3, inplace = True)
```

```python
df.info()
```

Dapat dilihat setelah didrop baris dengan minimal 3 kolom kosong, data menjadi hanya 110 entry, yang sebelumnya data memilki 210 entry

```python
df.isnull().sum()
```

```python
#Persentase missing value
(df.isnull().sum()/len(df)).to_frame('persentase missing')
```

dari hasil persentase missing value di atas, terlihat bahwa variabel "Bidang wirausaha:", dan "Tahun saudara mulai berwirausaha:" memiliki persentase missing value lebih dari 85%, diasumsikan data dari variabel tersebut tidak terlalu signifikan berarti

```python
#Exclude variabel yang tidak diperlukan
df.drop("Timestamp" , axis = 1, inplace = True) #Drop karena hanya berisi waktu pengisian
df.drop(["Bidang wirausaha:", "Tahun saudara mulai berwirausaha:"], axis = 1, inplace = True) #Drop karena persentase missing value yang cukup besar (> 85%)
```

```python
df.dtypes
```

Terdapat beberapa variabel bertipe objek, selanjutnya akan diubah menjadi tipe data kategori, begitu juga dengan variabel Tahun Masuk dan Tahun Wisuda akan diubah menjadi tipe data integer

ada keanehan pada data "Tahun masuk saat anda melanjutkan studi setelah lulus dari Prodi Matematika:" yang bertipe object, akan dilihat terlebih dahulu bagaimana isi dari variabel tersebut

```python
#Melihat isi dari variabel 'Tahun masuk saat anda melanjutkan studi setelah lulus dari Prodi Matematika:'
df['Tahun masuk saat anda melanjutkan studi setelah lulus dari Prodi Matematika:'].unique()
```

```python
#Replace data
df['Tahun masuk saat anda melanjutkan studi setelah lulus dari Prodi Matematika:'] = df['Tahun masuk saat anda melanjutkan studi setelah lulus dari Prodi Matematika:'].replace('S2 2008 dan S3 2020', 2008)
df['Tahun masuk saat anda melanjutkan studi setelah lulus dari Prodi Matematika:'] = df['Tahun masuk saat anda melanjutkan studi setelah lulus dari Prodi Matematika:'].replace('-', np.nan)
```

```python
df['Tahun masuk saat anda melanjutkan studi setelah lulus dari Prodi Matematika:'].unique()
```

```python
#Ubah ke tipe data yang benar
df['Status saat ini:'] = df['Status saat ini:'].astype('category')
df['Waktu Mendapatkan Pekerjaan:'] = df['Waktu Mendapatkan Pekerjaan:'].astype('category')
df['Posisi anda saat ini di tempat bekerja saat Ini:'] = df['Posisi anda saat ini di tempat bekerja saat Ini:'].astype('category')
df['Nama Universitas dan Kota tempat melanjutkan Studi setelah S1'] = df['Nama Universitas dan Kota tempat melanjutkan Studi setelah S1'].astype('category')
df['Nama jurusan/program studi setelah S1'] = df['Nama jurusan/program studi setelah S1'].astype('category')
df['Tahun Masuk'] = df['Tahun Masuk'].astype('int64')
df['Tahun Wisuda'] = df['Tahun Wisuda'].astype('int64')
```

```python
df.dtypes
```

```python
df[df['Posisi anda saat ini di tempat bekerja saat Ini:'].notna()]
```

```python
#Melihat duplikat pada data
df.duplicated().sum()
```

```python
#Menambahkan variabel lama kuliah pada data
df['lama kuliah'] = df['Tahun Wisuda'] - df['Tahun Masuk']
```

```python
df.rename(columns = {"Status saat ini:": "Status", "Perkiraan rata-rata penghasilan (take-home-pay) anda per-bulan saat ini (dalam juta rupiah):": "Penghasilan(dalam juta)"}, inplace = True)
```
